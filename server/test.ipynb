```python
# Use the sounddevice module
# http://python-sounddevice.readthedocs.io/en/0.3.10/

import numpy as np
import sounddevice as sd
import time

# Samples per second
rate=16000

# Frequency / pitch
freq_hz = 440.0

# Duration
duration_s = 5.0

# Attenuation so the sound is reasonable
atten = 0.3

# NumpPy magic to calculate the waveform
each_sample_number = np.arange(duration_s * rate)
waveform = np.sin(2 * np.pi * each_sample_number * freq_hz / rate)
waveform_quiet = waveform * atten
print(waveform_quiet)
```

```python
# Play the waveform out the speakers
sd.play(audio, rate)
time.sleep(duration_s)
sd.stop()
```
